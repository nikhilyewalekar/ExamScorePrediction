```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df = pd.read_csv("Exam_Score_Prediction.csv")
df.head()
```

```python
x = df.drop(columns = ['exam_score'])
y = df.exam_score
encoded = pd.get_dummies(x, dtype = int)
encoded.drop(columns = ['student_id'], inplace = True)
encoded
```

```python
x_train,x_test,y_train,y_test = train_test_split(encoded, y,test_size=0.2, random_state=10)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
y_train_predict = linear_model.predict(x_train)

mse = mean_squared_error(y_train, y_train_predict)
print("*"*30)
print("mse : ", mse)
print("*"*30)

r2score = r2_score(y_train, y_train_predict)

print("r2score : ", r2score)
print("*"*30)
```

```python
y_test_predict = linear_model.predict(x_test)

mse = mean_squared_error(y_test, y_test_predict)
print("*"*30)
print("mse : ", mse)
print("*"*30)

r2score = r2_score(y_test, y_test_predict)
print("r2score : ", r2score)
print("*"*30)
```
